## Problems

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
%%sql
select count(ID) as crimes_recorded
from CHICAGO_CRIME_DATA;
```

### Problem 2

##### List community area names and numbers with per capita income less than 11000.


```python
%%sql
-- Find all column names
PRAGMA table_info(CENSUS_DATA);
```

```python
%%sql
select community_area_name, community_area_number, PER_CAPITA_INCOME
from CENSUS_DATA
where PER_CAPITA_INCOME < 11000;
```

### Problem 3

##### List all case numbers for crimes involving minors?(children are not considered minors for the purposes of crime analysis) 


```python
%%sql

-- Selecting the case numbers for crimes involving minors
SELECT 
    CASE_NUMBER, -- Selecting the CASE_NUMBER column to list the case numbers
    PRIMARY_TYPE,
    DESCRIPTION
FROM 
    CHICAGO_CRIME_DATA -- Specifying the table to select the data from
WHERE 
    -- Filtering the descriptions to find crimes involving minors
    DESCRIPTION LIKE '%MINOR%' OR -- Checking for the term 'MINOR' in the DESCRIPTION
    DESCRIPTION LIKE '%CHILD%' OR -- Checking for the term 'CHILD' in the DESCRIPTION
    DESCRIPTION LIKE '%JUVENILE%' OR -- Checking for the term 'JUVENILE' in the DESCRIPTION
    PRIMARY_TYPE LIKE '%MINOR%' OR
    PRIMARY_TYPE LIKE '%CHILD%' OR
    PRIMARY_TYPE LIKE '%JUVENILE%';
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
%%sql
select CASE_NUMBER, PRIMARY_TYPE, DESCRIPTION
from CHICAGO_CRIME_DATA
where 
    PRIMARY_TYPE like '%kidnap%' and 
    (DESCRIPTION like '%child%' or
    DESCRIPTION like '%minor%' or
    DESCRIPTION like '%JUVENILE%');
```

### Problem 5

##### List the kind of crimes that were recorded at schools. (No repetitions)


```python
%%sql
select distinct primary_type
from CHICAGO_CRIME_DATA
where location_description like '%school%'
```

### Problem 6

##### List the type of schools along with the average safety score for each type.


```python
%%sql

-- Find all column names
PRAGMA table_info(CHICAGO_PUBLIC_SCHOOLS_DATA);
```

```python
%%sql

select [Elementary, Middle, or High School], avg(safety_score) as average_safety_score
from CHICAGO_PUBLIC_SCHOOLS_DATA
group by [Elementary, Middle, or High School]
order by average_safety_score asc;
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line


```python
%%sql

-- Find all column names
PRAGMA table_info(CENSUS_DATA);
```

```python
%%sql
select community_area_name, PERCENT_HOUSEHOLDS_BELOW_POVERTY
from CENSUS_DATA
order by PERCENT_HOUSEHOLDS_BELOW_POVERTY desc limit 5;
```

### Problem 8

##### Which community area is most crime prone? Display the coumminty area number only.


```python
%%sql

-- Find all column names
PRAGMA table_info(CHICAGO_CRIME_DATA);
```

```python
%%sql
select CE.community_area_number, CE.community_area_name, count(CR.id) as crime_count
from CENSUS_DATA CE, CHICAGO_CRIME_DATA CR
where CE.COMMUNITY_AREA_NUMBER = CR.COMMUNITY_AREA_NUMBER
group by CE.community_area_name
order by crime_count desc limit 1;
```

```python
%%sql

-- Selecting the community area number with the highest number of crimes
SELECT 
    CR.COMMUNITY_AREA_NUMBER
FROM 
    CHICAGO_CRIME_DATA CR
GROUP BY 
    CR.COMMUNITY_AREA_NUMBER
ORDER BY 
    COUNT(CR.CASE_NUMBER) DESC
LIMIT 1;
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
%%sql

select COMMUNITY_AREA_NAME, HARDSHIP_INDEX
from CENSUS_DATA
where HARDSHIP_INDEX IN (select max(HARDSHIP_INDEX)
                       from CENSUS_DATA);
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
%%sql
select CE.community_area_name
from CENSUS_DATA CE
where CE.community_area_number = (select CR.community_area_number
                                  from CHICAGO_CRIME_DATA CR
                                  group by CR.community_area_number
                                  order by count(cr.case_number) desc limit 1);
```
