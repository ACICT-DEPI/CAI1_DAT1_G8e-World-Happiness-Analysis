````python
"""
This code snippet creates a line plot to visualize the average automobile sales per year for different vehicle types.
It groups the data by year and vehicle type, calculates the mean sales, and plots the results to analyze the sales trends
for different vehicle types during recession periods.
"""

# Group the data by 'Year' and 'Vehicle_Type', then calculate the mean 'Automobile_Sales' for each group.
# This provides the average automobile sales per year for each vehicle type.
# The 'as_index=False' parameter prevents the grouped columns from becoming the index of the resulting DataFrame.
df_Mline = df.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()
print(df_Mline.head())  # Print the first few rows to inspect the grouped data.

# Set the 'Year' column as the index of the DataFrame.
# This makes it easier to plot the data with years on the x-axis.
# The 'inplace=True' parameter modifies the DataFrame in place without creating a new one.
df_Mline.set_index('Year', inplace=True)
print(df_Mline.head())  # Print the first few rows to inspect the DataFrame after setting the index.

# Group the data again by 'Vehicle_Type' to prepare it for plotting.
# This step ensures that the plot will have separate lines for each vehicle type.
# The 'groupby' method is used to group the DataFrame by 'Vehicle_Type'.
# The result is a groupby object that can be directly used for plotting.
df_Mline = df_Mline.groupby('Vehicle_Type')['Automobile_Sales']
print(df_Mline.head())  # Print the first few rows to inspect the DataFrame after grouping by 'Vehicle_Type'.

# Plot the data using the 'plot' method with 'line' as the plot type.
# The 'kind' parameter specifies the type of plot to be drawn, in this case, a line plot.
# Because the data is grouped by 'Vehicle_Type', Pandas will automatically create a separate line for each vehicle type.
df_Mline.plot(kind='line')

# Set the label for the x-axis.
# Replace '..............' with an appropriate label, such as 'Year'.
plt.xlabel('Year')

# Set the label for the y-axis.
# Replace '............' with an appropriate label, such as 'Average Automobile Sales'.
plt.ylabel('Average Automobile Sales')

# Set the title of the plot.
# The title should describe the content of the plot.
plt.title('Sales Trend Vehicle-wise during Recession')

# Add a legend to the plot for clarity.
# The legend helps identify different lines in the plot, corresponding to different vehicle types.
plt.legend()

# Display the plot.
plt.show()
````

````python
df.columns
# Columns to work with: 'Automobile_Sales', 'Recession', 'Vehicle_Type'
````

````python
"""
This code snippet creates a bar chart using Seaborn to compare the sales trend per vehicle type during a recession period and a non-recession period.
It groups the data by 'Recession', calculates the mean automobile sales, and plots the results.
"""

# Group the data by 'Recession' and calculate the mean 'Automobile_Sales' for each group.
# This provides the average automobile sales for recession and non-recession periods.
new_df = df.groupby('Recession')['Automobile_Sales'].mean().reset_index()
print(new_df.head())

# Create a new figure for the plot with specified dimensions.
plt.figure(figsize=(10, 6))

# Create the bar chart using Seaborn.
# 'x' specifies the column to be used for the x-axis.
# 'y' specifies the column to be used for the y-axis.
# 'hue' specifies the column to be used for color encoding.
# 'data' specifies the DataFrame to be used.
sns.barplot(x='Recession', y='Automobile_Sales', hue='Recession', data=new_df)

# Set the label for the x-axis.
plt.xlabel('Recession Period')

# Set the label for the y-axis.
plt.ylabel('Average Automobile Sales')

# Set the title of the plot.
plt.title('Average Automobile Sales during Recession and Non-Recession')

# Set the x-axis tick labels to 'Non-Recession' and 'Recession'.
plt.xticks(ticks=[0, 1], labels=['Non-Recession', 'Recession'])

# Display the plot.
plt.show()
````

````python
# Import necessary libraries
import folium
import pandas as pd
import requests
from io import BytesIO

# Define a function to download the GeoJSON file
def download_file(url, filename):
    response = requests.get(url)
    with open(filename, 'wb') as f:
        f.write(response.content)

# Define the path to the GeoJSON file and download it
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/us-states.json'
filename = 'us-states.json'
download_file(url, filename)

# Load the dataset into a DataFrame
csv_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv'
response = requests.get(csv_url)
data = pd.read_csv(BytesIO(response.content))

# Filter the data for the recession period and specific cities
recession_data = data[data['Recession'] == 1]

# Calculate the total sales by city during the recession period
sales_by_city = recession_data.groupby('City')['Automobile_Sales'].sum().reset_index()

# Create a base map centered on the United States
map1 = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

# Ensure the city names match between the DataFrame and GeoJSON
sales_by_city.columns = ['name', 'Automobile_Sales']

# Add the GeoJSON layer
folium.Choropleth(
    geo_data=filename,
    data=sales_by_city,
    columns=['name', 'Automobile_Sales'],
    key_on='feature.properties.name',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Automobile Sales during Recession'
).add_to(map1)

# Add tooltips to the GeoJSON layer
folium.GeoJson(
    filename,
    name="Sales",
    tooltip=folium.GeoJsonTooltip(fields=['name'], aliases=['City:'])
).add_to(map1)

# Display the map
map1
````
